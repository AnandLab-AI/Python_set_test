```python
#cenario: A system checks if a user is eligible to vote based on their age.
#Write logic to ask the user for their age and determine if they are eligible to vote based on whether they are 18 or older.

name=input("Enter your name here:")
age=int(input("Enter your age here:"))
if(age>=18):
    print(f"Your Name:{name}\n"
          f"Your Age:{age}\n"
          f"You are eligible for voting")
else:
    print(f"Your Name:{name}\n"
          f"Your Age:{age}\n"
          f"You are not eligible for vote")
```

```python
#6.	Scenario: A grading system determines whether a student has passed or failed based on their score.
# Write logic to check if a student has passed a subject by scoring at least 40 marks.

Tamil=int(input("Enter your Tamil mark here:"))
English=int(input("Enter your English mark here:"))
Maths=int(input("Enter your Maths mark here:"))
Science=int(input("Enter your Science mark here:"))
Social_Science=int(input("Enter your Social mark here:"))

L1=[("Tamil",Tamil),("English",English),("Maths",Maths),("Science",Science),("Social Science",Social_Science)]

for subj,mark in L1:
    if mark >= 40 :
        print (f"{subj}:{mark} - PASS ")
    else:
        print (f"{subj}:{mark} - FAIL ")
T1=[]
for subj,mark in L1:  
    T1.append(mark)
Total=sum(T1)
percentage=Total/500*100
print("============")
print(f"Your Total:{Total}\n"
      f"Percentage :{percentage}")
```

```python
#7.	Scenario: A retail #store offers a 20% discount if a customer’s total order exceeds $100. Write logic to calculate the final amount to be paid after applying the discount.

amt=int(input("Enter your total amount:"))
if amt >= 100:
    discount=amt*20/100
    final=amt-discount
    print(f"Your discount amount :{discount}\n"
          f"Your total amount is:{final}")
else:
    print("Your total amount is:",amt)
```

```python
#8.	Scenario: A banking system processes withdrawal requests and ensures the user has enough balance.
#Write logic to check if a user has enough balance before allowing a withdrawal and update the remaining balance accordingly.

Camt=int(input("Deposit amount:"))
Wamt=int(input("Witdraw amount:"))
if (Wamt <= Camt ):
    AvlBal=Camt-Wamt
    print(f"Your Withdraw amount:{Wamt}\n"
          f"Your Current balance:{AvlBal}")
else:
    print("No enough balance to withdraw")
```

```python
#Scenario: A calendar system verifies whether a given year is a leap year based on standard leap year rules.
#Write logic to determine whether a given year is a leap year

year = int(input("Enter year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
#10.	Scenario: A program filters out only even numbers from a given list.
# Write logic to extract and return only the even numbers from a list.

X1=[10,23,35,44,55,66,77,88,99,100]
EList=[]
for E1 in X1:
    if(E1%2 == 0):
        EList.append(E1)
print(f"Even number List :\n"
      f"{EList}")
    
```
