```python
#3.Write a Python function to check if a given string is a palindrome.
#Expected Output (for input "madam"):
#madam is a palindrome

def mypalindrome(mypa):
    mypa=mypa.lower().replace(" ","")
    
    if mypa==mypa[::-1]:
        print(f"Entered text {mypa} is palindrome")
    else:
        print(f"Entered text {mypa} is not palindrome")
        
mypa=input("Enter your text here:")    
mypalindrome(mypa)
```
